## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def transform_pascal_snake(text, case_type = None):
    
    if not(len(text)):
        return
    
    elif case_type == "S" or ("_" not in text and text[0].isupper() and not case_type):
        
        new_chars = ["_" + char.lower() if char.isupper() else char for char in text]
        new_chars[0] = new_chars[0].lstrip("_")
          
    elif case_type == "P" or (("_" in text or text[0].islower()) and not case_type):

        new_chars = [i[0].upper() + i[1:] for i in text.split("_")]
    
    return "".join(new_chars)  

inp = input("Enter text: ")
print(transform_pascal_snake(inp, "P"))
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def is_valid_date(inp_date):

    dd, mm, yyyy = map(int, inp_date.split("."))

    if dd < 1 or mm < 1 or mm > 12 or yyyy <= 0:
        return False

    is_leap_year = not yyyy % 4 and yyyy % 100 or not yyyy % 400
    month_days = (31, 28 + is_leap_year, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)

    if dd > month_days[mm - 1]:
        return False
    
    return True
    
inp_date = input("Enter date: ")
print(is_valid_date(inp_date))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_digit_simple(digit):

    is_simple = True
    
    if digit <= 1:
        is_simple = False
    else:
        for sub_digit in range(2, digit):
            if not digit % sub_digit:
                is_simple = False
                break
                
    return is_simple

inp = int(input("Enter number: "))
print(is_digit_simple(inp))
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
#== Функция для печати словаря
def dict_print(inp_dict):
    
    print(f"\n{"User_id":<10} {"First_Name":<15} {"Last_Name":<15} {"Age":<5}")
    print(f"{"=" * 7:<10} {"=" * 10:<15} {"=" * 9:<15} {"=" * 3:<5}")
    
    for k, i in inp_dict.items():
        print(f"{k:<10} {i[0]:<15} {i[1]:<15} {i[2]:<5}")

#== Основной модуль:
#== добавление пользователей с последующим выводом  
user_dict = dict()

while True:
    
    inp_data = input("Enter user's data as a string [FirstName LastName Age ID]: ").split()
    
    if not inp_data:
        break
    
    name, fam, age, us_id = inp_data
    
    age = int(age)
    us_id = int(us_id + "0" * (8 - len(us_id)))
    
    if us_id in user_dict:
        print(f"ID {us_id} already exists!")
        continue
    elif age < 18 or age > 60:
        print(f"{age} - Unacceptable age!")
        continue
    elif not name.isalpha() or not fam.isalpha():
        print("First/Last name contains invalid symbols!")
        continue

    user_dict[us_id] = name.capitalize(), fam.capitalize(), age
    
dict_print(user_dict)
```
