## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
input_number = input("Enter 5-digit number:")

mirror_number = input_number[0] + input_number[3:0:-1] + input_number[4]

print(f"{input_number} -> {mirror_number}")
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
days_before = int(input("Days before vacancy:"))

days_left = days_before // 7 * 2 

if days_before % 7 == 6:
    days_left += 1

print(f"\n{days_before} -> {days_left}")
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
length = int(input("Enter chocolate length:"))
width = int(input("Enter chocolate width:"))
piece_size = int(input("Enter piece size:"))

is_splittable = False

if piece_size % length == 0 or piece_size % width == 0:
    is_splittable = True

print(f"\n{length}, {width}, {piece_size} -> {is_splittable}")
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
input_data = input("Enter integer positive number from 1 to 3999:")

decimal_number = int(input_data)
roman_number = ''

decimal_roman_match = [
    (1000, "M"),
    (900, "CM"),
    (500, "D"), 
    (400, "CD"),
    (100, "C"),
    (90, "XC"),
    (50, "L"),
    (40, "XL"),
    (10, "X"),
    (9, "IX"),
    (5, "V"),
    (4, "IV"),
    (1, "I")
]

while decimal_number:
    
    for dec_num, rom_num in (decimal_roman_match):
        
        if decimal_number >= dec_num:
            
            rom_num_counter = decimal_number // dec_num
            decimal_number -= rom_num_counter * dec_num
            roman_number += rom_num_counter * rom_num
            
            break
        
print(f"{input_data} -> {roman_number}")
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
input_data = input("Enter some data: ").strip()

is_number = True
number_sign = ''
first_symbol = input_data[0]

if first_symbol == "e" or input_data[len(input_data)-1] in ".e-":
    is_number = False
elif input_data.count(".") > 1 or " " in input_data or ".-" in input_data:
    is_number = False
elif ".e" in input_data or "e." in input_data or "-e" in input_data:
    is_number = False
else:
    for symbol in input_data:
        if symbol not in "0123456789.e-":
            is_number = False
            break

if is_number == True: 
    if first_symbol == "-":
        number_sign = ", Is_Negative"
    else:
        number_sign = ", Is_Positive"
        
print(f"{input_data} -> {is_number}{number_sign}")
```
